```python
## Task 1: A list of dictionaries

transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]

# a) Filter valid transactions
valid_transactions = [
    transaction for transaction in transactions
    if transaction["price"] > 0 and transaction["quantity"] >= 0
]

# b) Add total field
for transaction in valid_transactions:
    transaction["total"] = transaction["price"] * transaction["quantity"]

# c) Group by region and compute total revenue
revenue_by_region = {}

for transaction in valid_transactions:
    region = transaction["region"]
    total = transaction["total"]
    
    if region not in revenue_by_region:
        revenue_by_region[region] = 0
    
    revenue_by_region[region] += total

# d) Find region with highest revenue
best_region = max(revenue_by_region, key=revenue_by_region.get)
best_value = revenue_by_region[best_region]

print("Valid transactions:")
for transaction in valid_transactions:
    print(transaction)

print("\nRevenue by region:")
print(revenue_by_region)

print(f"\nThe region with the highest total revenue is {best_region} with a value of {best_value}.")
```

### Explanation

First, invalid transactions were removed by excluding rows with a price equal to zero or a negative quantity.  
Then, a new field called `total` was created for each valid transaction.  
After that, the valid transactions were grouped by region and the total revenue was calculated for each region.  
Finally, the region with the highest total revenue was identified and printed using an f-string.

## Task 2: Creating a function

```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

def summarise_sales(sales):
    # total
    total = sum(sales)
    
    # average
    average = round(total / len(sales), 2)
    
    # best and worst day (index starts at 1)
    best_day = sales.index(max(sales)) + 1
    worst_day = sales.index(min(sales)) + 1
    
    # count above average
    above_average_count = 0
    for value in sales:
        if value > average:
            above_average_count += 1
    
    # return dictionary
    return {
        "total": total,
        "average": average,
        "best_day": best_day,
        "worst_day": worst_day,
        "above_average_count": above_average_count
    }

# call function
result = summarise_sales(sales)

print(result)
```

### Explanation

The function calculates key statistics from the sales data.  
First, it computes the total and the average.  
Then, it identifies the best and worst days based on maximum and minimum sales values.  
Finally, it counts how many days had sales above the average.  
All results are returned in a dictionary.

## Task 3: Loops

```python
students = [
    {"name": "Anna", "grade": 85, "attendance": 0.92},
    {"name": "Erik", "grade": 58, "attendance": 0.75},
    {"name": "Mia", "grade": 91, "attendance": 0.98},
    {"name": "Leo", "grade": 44, "attendance": 0.60},
    {"name": "Sara", "grade": 72, "attendance": 0.88},
    {"name": "Tom", "grade": 67, "attendance": 0.70},
    {"name": "Nina", "grade": 95, "attendance": 0.99},
    {"name": "Olli", "grade": 50, "attendance": 0.65},
]

# a) passing students
passing = []

for student in students:
    if student["grade"] >= 50:
        passing.append(student["name"])

print("Passing students:", passing)


# b) status messages
for student in students:
    name = student["name"]
    grade = student["grade"]
    
    if grade >= 80:
        print(f"{name}: Excellent")
    elif grade >= 50:
        print(f"{name}: Pass")
    else:
        print(f"{name}: Fail")


# c) attendance below 0.80
low_attendance_count = 0

for student in students:
    if student["attendance"] < 0.80:
        low_attendance_count += 1

print("\nStudents with attendance below 0.80:", low_attendance_count)


# d) highest grade WITHOUT max()
best_student = None
best_grade = 0

for student in students:
    if student["grade"] > best_grade:
        best_grade = student["grade"]
        best_student = student

print("\nBest student (without max):", best_student["name"], "-", best_grade)


# e) highest grade WITH max()
best_student_max = max(students, key=lambda x: x["grade"])

print("Best student (with max):", best_student_max["name"], "-", best_student_max["grade"])
```

### Explanation

This task uses loops and conditional statements to analyze student data.

First, a list of students who passed is created by checking if the grade is at least 50.  
Then, a message is printed for each student depending on their grade.  
After that, the number of students with attendance below 0.80 is counted.  
Finally, the student with the highest grade is found using both a loop and the max() function.

## Task 4: Class and methods

```python
class Portfolio:
    
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}
    
    # a) add asset
    def add_asset(self, ticker, shares, price_per_share):
        self.holdings[ticker] = {
            "shares": shares,
            "price_per_share": price_per_share
        }
    
    # b) total value
    def total_value(self):
        total = 0
        for asset in self.holdings.values():
            total += asset["shares"] * asset["price_per_share"]
        return total
    
    # c) most valuable asset
    def most_valuable(self):
        best_ticker = None
        best_value = 0
        
        for ticker, asset in self.holdings.items():
            value = asset["shares"] * asset["price_per_share"]
            
            if value > best_value:
                best_value = value
                best_ticker = ticker
        
        return best_ticker
    
    # d) string representation
    def __str__(self):
        return f"Portfolio of {self.owner} with {len(self.holdings)} assets. Total value: {self.total_value()}"

# create portfolio
p = Portfolio("Emilien")

# add assets
p.add_asset("AAPL", 10, 150)
p.add_asset("TSLA", 5, 700)
p.add_asset("GOOG", 2, 2000)

# test methods
print("Total value:", p.total_value())
print("Most valuable asset:", p.most_valuable())
print(p)
```

### Explanation

A class Portfolio was created to represent an investment portfolio.

The add_asset method allows adding or updating assets.  
The total_value method calculates the total portfolio value.  
The most_valuable method identifies the asset with the highest value.  
The __str__ method provides a readable summary of the portfolio.

## Task 5: Importing a module

```python
from datetime import datetime

orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]

# convert strings to date objects
order_dates = []

for order in orders:
    date_string = order[0]
    date_object = datetime.strptime(date_string, "%Y-%m-%d").date()
    order_dates.append(date_object)

# find earliest and latest date
earliest_order = min(order_dates)
latest_order = max(order_dates)

# compute difference in days
days_between = (latest_order - earliest_order).days

print("Earliest order:", earliest_order)
print("Latest order:", latest_order)
print("Days between earliest and latest order:", days_between)
```

### Explanation

The datetime module was used to convert each date string into a date object.  
Then, the earliest and latest order dates were identified using min() and max().  
Finally, the difference in days between these two dates was calculated.

## Task 6: Array Operations and Broadcasting

```python
import numpy as np

revenue = np.array([
    [210, 230, 190, 250],
    [220, 260, 200, 270],
    [250, 290, 230, 310]
])

# a) total annual revenue (sum of rows)
annual_totals = np.sum(revenue, axis=1)
print("Annual totals:", annual_totals)


# b) mean quarterly revenue (column means)
quarterly_means = np.mean(revenue, axis=0)
print("Quarterly means:", quarterly_means)


# c) apply growth to 2024 (last row)
growth = np.array([1.05, 1.03, 1.08, 1.02])

adjusted_2024 = revenue[2] * growth
print("Adjusted 2024 revenue:", adjusted_2024)


# d) values greater than 250
high_values = revenue[revenue > 250]
print("Values greater than 250:", high_values)
```

### Explanation

NumPy was used to perform efficient array operations.

First, row sums were computed to get annual revenue.  
Then, column means were calculated to find average quarterly revenue.  
Next, broadcasting was used to apply growth factors to the 2024 data.  
Finally, boolean indexing was used to extract values greater than 250.

## Task 7: Pandas

```python
import pandas as pd

# create DataFrame
data = {
    "employee": ["Anna", "Erik", "Mia", "Leo", "Sara", "Tom"],
    "department": ["Finance", "Marketing", "Finance", "IT", "Marketing", "IT"],
    "salary": [4500, 4000, 4800, 5200, 3900, 5100],
    "years_exp": [5, 3, 7, 10, 2, 8],
    "performance_score": [4.2, 3.8, 4.5, 4.0, 3.5, 4.9]
}

df = pd.DataFrame(data)

# a) salary per year of experience
df["salary_per_year_exp"] = (df["salary"] / df["years_exp"]).round(0)

print("DataFrame with new column:")
print(df)


# b) sort by performance_score
top_3 = df.sort_values(by="performance_score", ascending=False).head(3)

print("\nTop 3 employees by performance:")
print(top_3)


# c) finance employees with score > 4.0
finance_high_perf = df.loc[
    (df["department"] == "Finance") & (df["performance_score"] > 4.0)
]

print("\nFinance employees with high performance:")
print(finance_high_perf)


# d) rename column
df = df.rename(columns={"years_exp": "experience_years"})

print("\nDataFrame after renaming column:")
print(df)


# e) select salary and performance for first 4 employees
subset = df.iloc[:4][["salary", "performance_score"]]

print("\nSubset of first 4 employees:")
print(subset)
```

### Explanation

Pandas was used to create and manipulate a DataFrame.

First, a new column was created by dividing salary by years of experience.  
Then, the DataFrame was sorted by performance score to find the top employees.  
Next, filtering with .loc was used to select specific rows.  
After that, a column was renamed using the rename() method.  
Finally, .iloc was used to select specific rows and columns.

## Task 8: More Pandas

```python
import numpy as np
import pandas as pd

# generate data
rng = np.random.default_rng(seed=123)

days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)

# a) create DataFrame
df = pd.DataFrame({
    "date": days,
    "visitors": visitors,
    "conversion_rate": conversion_rate,
    "avg_order_value": avg_order_value
})

df = df.set_index("date")

print("Initial DataFrame:")
print(df.head())


# b) compute new columns
df["conversions"] = (df["visitors"] * df["conversion_rate"]).round(0).astype(int)

df["daily_revenue"] = df["conversions"] * df["avg_order_value"]

print("\nDataFrame with new columns:")
print(df.head())
```

### Explanation

A DataFrame was created using the generated data and the date column was set as the index.

Then, the number of conversions was calculated by multiplying visitors by the conversion rate and rounding to the nearest integer.

Finally, daily revenue was computed by multiplying conversions by the average order value.
